```julia
delete!(hbmass, [636, 637, 638])
template = filter!(row -> !(row.Code_Sujet in ["640", "BEJO674", "QUJI740"]), template)
```

```julia
anemie = select!(anemie, Not([:Age_mois, :Sexe_F_Fille_G_Garçon, :Poids, :Ville_eval, :Taille, :Age_categorie, :Age_categorie_binaire]))
cardio = select!(cardio, Not([:Age_mois, :Sexe_F_Fille_G_Garçon, :Poids, :Ville_eval, :Taille, :Age_categorie, :Age_categorie_binaire]))
nfs_visco = select!(nfs_visco, Not([:Age_mois, :Sexe_F_Fille_G_Garçon, :Poids, :Ville_eval, :Taille, :Age_categorie, :Age_categorie_binaire]))
hbmass = select!(hbmass, Not([:Age_mois, :Sexe_F_Fille_G_Garçon, :Poids, :Ville_eval, :Taille, :Age_categorie, :Age_categorie_binaire]))
```

```julia
data = leftjoin(template, anemie, on = :Code_Sujet)
```

```julia
data = leftjoin(data, cardio, on = :Code_Sujet)
```

```julia
data = leftjoin(data, nfs_visco, on = :Code_Sujet)
```

```julia
sum(skipmissing(data."Hb mass (g)") .!= skipmissing(hbmass."Hb mass (g)"))
sum(skipmissing(data."Hb mass/kg (g / kg)") .!= skipmissing(hbmass."Hb mass/kg (g / kg)"))
```

```julia
hbmass = select!(hbmass, Not([:"Hb mass (g)", :"Hb mass/kg (g / kg)"]))
```

```julia
data = leftjoin(data, hbmass, on = :Code_Sujet)
```
